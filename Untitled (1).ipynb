```python
def num_of_factors(num):
    factors=[i for i in range(1,num+1) if num%i==0]
    print(factors)
    return len(factors)
num=int(input('enter the integer:'))
print(num_of_factors(num))
```

```python
def root(x,n=2):
    return(x)**(1/n)
num=int(input('enter the number:'))
print('the root is',root(num))
```
